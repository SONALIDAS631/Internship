```python
import requests
from bs4 import BeautifulSoup

base_url = "https://www.shine.com"
url = f"{base_url}/job-search/data-analyst-jobs-in-bangalore"

response = requests.get(url)

if response.status_code == 200:
  
    soup = BeautifulSoup(response.content, "html.parser")

    job_listings = soup.find_all("li", class_="search_listing")

    job_data = []
    for listing in job_listings:
        job_title = listing.find("h3", class_="job_title").text.strip()
        job_location = listing.find("span", class_="job_location").text.strip()
        
        job_data.append({
            "Job Title": job_title,
            "Job Location": job_location
        })
   
    for job in job_data:
        print("Job Title:", job["Job Title"])
        print("Job Location:", job["Job Location"])
        print("-" * 30)
else:
    print("Failed to retrieve the webpage")
```

```python
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from bs4 import BeautifulSoup
import pandas as pd
import time

driver_path = "/path/to/chromedriver"
driver = webdriver.Chrome(executable_path=driver_path)

base_url = "https://www.shine.com"
url = f"{base_url}/job-search/data-analyst-jobs-in-bangalore"

driver.get(url)

job_title_input = driver.find_element_by_id("id_q")
location_input = driver.find_element_by_id("id_l")
search_button = driver.find_element_by_id("id_btn")

job_title_input.send_keys("Data Analyst")
location_input.send_keys("Bangalore")
search_button.click()

time.sleep(3)

soup = BeautifulSoup(driver.page_source, "html.parser")
job_listings = soup.find_all("li", class_="search_listing")

job_data = []
for listing in job_listings[:10]:  # Limit to the first 10 results
    job_title = listing.find("h3", class_="job_title").text.strip()
    job_location = listing.find("span", class_="job_location").text.strip()
    company_name = listing.find("span", class_="company_name").text.strip()
    experience_required = listing.find("li", class_="wexp").text.strip()

    job_data.append({
        "Job Title": job_title,
        "Job Location": job_location,
        "Company Name": company_name,
        "Experience Required": experience_required
    })

driver.quit()

df = pd.DataFrame(job_data)

print(df)
```
